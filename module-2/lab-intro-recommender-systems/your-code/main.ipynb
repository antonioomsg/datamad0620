For this lab, we are going to be putting together a user similarity based recommender system in a step-by-step fashion. Our data set contains customer grocery purchases, and we will use similar purchase behavior to inform our recommender system. Our recommender system will generate 5 recommendations for each customer based on the purchases they have made.

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('../data/customer_product_sales.csv')
```

## Step 1: Create a data frame that contains the total quantity of each product purchased by each customer.

You will need to group by CustomerID and ProductName and then sum the Quantity field.

```python
total_per_customer= data.groupby(["CustomerID","ProductName"]).agg(Suma=("Quantity","sum"))
total_per_customer
```

## Step 2: Use the `pivot_table` method to create a product by customer matrix.

The rows of the matrix should represent the products, the columns should represent the customers, and the values should be the quantities of each product purchased by each customer. You will also need to replace nulls with zeros, which you can do using the `fillna` method.

```python
table = pd.pivot_table(data, values='Quantity', index=['ProductName'], columns=['CustomerID'], aggfunc=np.sum, fill_value=0)
table
```

## Step 3: Create a customer similarity matrix using `squareform` and `pdist`. For the distance metric, choose "euclidean."

```python
from scipy.spatial.distance import pdist, squareform

distances = pdist(table.T, 'euclidean')
print(distances)
print("-----Squareform-----")
print(squareform(distances))
```

```python
distances = pd.DataFrame(1/(1 + squareform(pdist(table.T, 'euclidean'))), 
                         index=table.columns, columns=table.columns)

distances
```

## Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.

```python
similarities = distances[33].sort_values(ascending=False)[1:6]
listcust = list(similarities.index)
listcust
```

## Step 5: From the data frame you created in Step 1, select the records for the list of similar CustomerIDs you obtained in Step 4.

```python
#simil_customers = similarities.axes
#country=[264,3535,3317,2503,3305]
#usa3= total_per_customer[(total_per_customer["CustomerID"].isin(country))]
total_buys_rec= total_per_customer.loc[listcust]
total_buys_rec
```

## Step 6: Aggregate those customer purchase records by ProductName, sum the Quantity field, and then rank them in descending order by quantity.

This will give you the total number of each product purchased by the 5 most similar customers to the customer you selected in order from most purchased to least.

```python
total_buys_rec.index = total_buys_rec.index.droplevel("CustomerID")
```

```python
total_buys_rec
index_by_product = total_buys_rec.reset_index()
total_buys_rec =index_by_product.groupby("ProductName").sum().sort_values(by='Suma',ascending=False)
total_buys_rec
```

## Step 7: Filter the list for products that the chosen customer has not yet purchased and then recommend the top 5 products with the highest quantities that are left.

- Merge the ranked products data frame with the customer product matrix on the ProductName field.
- Filter for records where the chosen customer has not purchased the product.
- Show the top 5 results.

```python
#customer_chosen= table.T
#customer_chosen= customer_chosen.reset_index()
#customer_chosen.iloc(1)

total_buys_rec= total_buys_rec.merge(table, left_on = "ProductName", right_on = "ProductName")
total_buys_rec_33= total_buys_rec[["Suma",33]]
```

```python
limit=0
total_buys_rec_33 = total_buys_rec_33[total_buys_rec_33[33]<=limit]
#soy demasiado fan de esta formulita :)
lista_33= total_buys_rec_33.sort_values(by='Suma',ascending=False).head(5)
lista_33_products = list(lista_33.index)
lista_33_products
```

## Step 8: Now that we have generated product recommendations for a single user, put the pieces together and iterate over a list of all CustomerIDs.

- Create an empty dictionary that will hold the recommendations for all customers.
- Create a list of unique CustomerIDs to iterate over.
- Iterate over the customer list performing steps 4 through 7 for each and appending the results of each iteration to the dictionary you created.

```python
rec_customers={}
simil_customers={}

for i in data.CustomerID:
    similarities_total = distances[i].sort_values(ascending=False)[1:6]
    listcust_total = list(similarities_total.index)
    simil_customers[i]=listcust_total
```

```python
for x,y in simil_customers.items():
    total_buys_rec_all = total_per_customer.loc[y]
    total_buys_rec_all.index = total_buys_rec_all.index.droplevel("CustomerID")
    index_by_product_all = total_buys_rec_all.reset_index()
    total_buys_rec_all =index_by_product_all.groupby("ProductName").sum().sort_values(by='Suma',ascending=False)
    total_buys_rec_all= total_buys_rec_all.merge(table, left_on = "ProductName", right_on = "ProductName")
    total_buys_rec_all_total= total_buys_rec_all[["Suma",x]]
    total_buys_rec_all_total = total_buys_rec_all_total[total_buys_rec_all_total[x]<=limit]
    lista_all= total_buys_rec_all_total.sort_values(by='Suma',ascending=False).head(5)
    lista_all_index= list(lista_all.index)
    rec_customers[x]=lista_all_index
    
# cuando se abren las recs para mercadona jajajaj?
```

```python
rec_customers
```

##  Step 9: Store the results in a Pandas data frame. The data frame should a column for Customer ID and then a column for each of the 5 product recommendations for each customer.

```python
final_df = pd.DataFrame([[key, rec_customers[key]] for key in rec_customers.keys()], columns=['id', 'products'])
final_df
tags = final_df['products'].apply(pd.Series)
tags = tags.rename(columns = lambda x : 'Product_' + str(x))
final_df= pd.concat([final_df[:], tags[:]], axis=1)
final_df
```
